```python
import pandas as pd
user_json = 'Datasets/IDEA/dataset/users.csv'

user = pd.read_csv(user_json)
print(user.head())

# load the data as dictionary
user_dict = user.to_dict(orient='records')
print(user_dict[0])
```

```python
import tqdm

# read data
bookmarks = pd.read_csv('Datasets/IDEA/dataset/bookmarks.csv')
favourites = pd.read_csv('Datasets/IDEA/dataset/favourites.csv')
interactions = pd.read_json('Datasets/IDEA/dataset/interactions.json')
ratings = pd.read_csv('Datasets/IDEA/dataset/ratings.csv')
recommendations = pd.read_json('Datasets/IDEA/dataset/recommendations.json')
surveys = pd.read_csv('Datasets/IDEA/dataset/surveys.csv')
views = pd.read_json('Datasets/IDEA/dataset/views.json')


# add the interactions data to the user data, according to the informfully_id(user_id)
for user in tqdm.tqdm(user_dict, total=len(user_dict)):
    user['bookmarks'] = bookmarks[bookmarks['informfully_id'] == user['informfully_id']].to_dict(orient='records')
    user['favourites'] = favourites[favourites['informfully_id'] == user['informfully_id']].to_dict(orient='records')
    user['interactions'] = interactions[interactions['informfully_id'] == user['informfully_id']].to_dict(orient='records')
    user['ratings'] = ratings[ratings['informfully_id'] == user['informfully_id']].to_dict(orient='records')
    user['recommendations'] = recommendations[recommendations['informfully_id'] == user['informfully_id']].to_dict(orient='records')
    user['surveys'] = surveys[surveys['informfully_id'] == user['informfully_id']].to_dict(orient='records')
    user['views'] = views[views['informfully_id'] == user['informfully_id']].to_dict(orient='records')
    del user['Unnamed: 0']
    for bookmark in user['bookmarks']:
        del bookmark['informfully_id']
        # convert the timestamp to string
        for key, value in bookmark.items():
            if isinstance(value, pd.Timestamp):
                bookmark[key] = value.isoformat()
    
    for favourite in user['favourites']:
        del favourite['informfully_id']
        # convert the timestamp to string
        for key, value in favourite.items():
            if isinstance(value, pd.Timestamp):
                favourite[key] = value.isoformat()
                
    for interaction in user['interactions']:
        del interaction['informfully_id']
        # convert the timestamp to string
        for key, value in interaction.items():
            if isinstance(value, pd.Timestamp):
                interaction[key] = value.isoformat()
    
    for rating in user['ratings']:
        del rating['informfully_id']
        # convert the timestamp to string
        for key, value in rating.items():
            if isinstance(value, pd.Timestamp):
                rating[key] = value.isoformat()
    
    for recommendation in user['recommendations']:
        del recommendation['informfully_id']
        # convert the timestamp to string
        for key, value in recommendation.items():
            if isinstance(value, pd.Timestamp):
                recommendation[key] = value.isoformat()
    
    for survey in user['surveys']:
        del survey['informfully_id']
        # convert the timestamp to string
        for key, value in survey.items():
            if isinstance(value, pd.Timestamp):
                survey[key] = value.isoformat()
    
    for view in user['views']:
        
        del view['informfully_id']
        # convert the timestamp to string
        for key, value in view.items():
            if isinstance(value, pd.Timestamp):
                view[key] = value.isoformat()
                
    # check if there is any NaN value in the data
    for key, value in user.items():
        if isinstance(value, list):
            for item in value:
                for k, v in item.items():
                    if pd.isna(v):
                        item[k] = None
        else:
            if pd.isna(value):
                user[key] = None

print(user_dict[0])

# export the data to json
with open('Datasets/IDEA/dataset/user_new.json', 'w') as file:
    json.dump(user_dict, file, indent=4)
```
